```python
# Histogram
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("onlinefoods.csv")

plt.hist(df['Pin code'])
plt.xlabel('Pin code')
plt.ylabel('Frequency')
plt.title('Distribution of Pin code')
plt.show()
```

```python
# Pie chart
import matplotlib.pyplot as plt
df['Marital Status'].value_counts().plot(kind='pie')
```

```python
# prompt: pie chart

import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("onlinefoods.csv")

# Pie chart
df['Monthly Income'].value_counts().plot(kind='pie')
plt.title('Monthly Income Distribution')
plt.show()
```

```python
# prompt: line

import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("onlinefoods.csv")

# Line chart
plt.plot(df['latitude'], df['longitude'])
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.title('latitude vs logitude')
plt.show()
```

```python
# prompt: scatter plot

# Scatter plot
plt.scatter(df['Age'], df['Monthly Income'])
plt.xlabel('Age')
plt.ylabel('Monthly Income')
plt.title('Age vs Monthly Income')
plt.show()
```

```python
# prompt: multi-series

import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
data = {
    'Year': [2010, 2011, 2012, 2013, 2014],
    'Sales_A': [100, 120, 150, 130, 160],
    'Sales_B': [80, 90, 110, 100, 120]
}
df = pd.DataFrame(data)

# Multi-series line chart
plt.plot(df['Year'], df['Sales_A'], label='Sales A')
plt.plot(df['Year'], df['Sales_B'], label='Sales B')

plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Trend Over Time')
plt.legend()
plt.show()
```
